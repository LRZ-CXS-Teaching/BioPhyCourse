This notebook provides the solution to Lammps Hello World example.

```python
from lammps import lammps
lmp = lammps()
```

ToDO: Define a 3D block type atomistic system of size 20x20x20 (lj unit) and having FCC latice, conisder lattice constant of 0.8442. Each atom has mass of 1.0.

```python
lmp.command("units  lj")
lmp.command("atom_style	atomic")

lmp.command("lattice         fcc 0.8442")
lmp.command("region          box block 0 20 0 20 0 20")
lmp.command("create_box	    1 box")
lmp.command("create_atoms    1 box")
lmp.command("mass            1 1.0")
```

ToDo: Initialize velocities for all atoms at reduced temperature 1.44. Atoms interact with LJ potential having cut-off radius 2.5. 

```python
lmp.command("velocity     all create 1.44 87287 loop geom")

lmp.command("pair_style      lj/cut 2.5")
lmp.command("pair_coeff      1 1 1.0 1.0 2.5")
```

ToDo: Update neighbor list every 5 steps, and neighbor list skin distance of 0.3. 

```python
lmp.command("neighbor        0.3 bin")
lmp.command("neigh_modify    delay 5")
```

Apply constant energy ensemble to all atoms for verlet time integration.  

```python
lmp.command("fix             1 all nve")
```

ToDo: print thermodiynamical property every 10 steps, and dump position and velocity of every atom every 10 steps. Run for 100 steps.

```python
lmp.command("thermo          10")
lmp.command ("dump           1 all custom 10 dump.lj id type x y z vx vy vz")
lmp.command("run             100")
```
